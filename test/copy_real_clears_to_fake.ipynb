```python
import os

from dotenv import load_dotenv

import sheets
from constants import CLEARS_PAGE_NAME, CLD_PAGE_NAME

load_dotenv()
```

```python
gc = sheets.get_gspread_client()
```

```python
def sync_worksheet(source_wks, target_sh):
    title = source_wks.title

    existing = target_sh.worksheet(title)
    target_sh.del_worksheet(existing)

    copy_data = source_wks.copy_to(target_sh.id)
    new_id = str(copy_data['sheetId'])

    for wks in target_sh.worksheets():
        if str(wks.id) == new_id:
            wks.update_title(title)
            break
```

```python
real_clears_sheet_id = os.environ.get('_REAL_CLEARS_SHEET_ID')
fake_clears_sheet_id = os.environ.get('_FAKE_CLEARS_SHEET_ID')
source_sh = gc.open_by_key(real_clears_sheet_id)
dest_sh = gc.open_by_key(fake_clears_sheet_id)
clears = source_sh.worksheet(CLEARS_PAGE_NAME)
cld = source_sh.worksheet(CLD_PAGE_NAME)

sync_worksheet(clears, dest_sh)
sync_worksheet(cld, dest_sh)
```
