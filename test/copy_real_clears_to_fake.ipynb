```python
import os

from dotenv import load_dotenv

import sheets

load_dotenv()
```

```python
gc = sheets.authenticate_gspread()
```

```python
real_clears_sheet_id = os.environ.get('_REAL_CLEARS_SHEET_ID')
fake_clears_sheet_id = os.environ.get('_FAKE_CLEARS_SHEET_ID')
clears_page_name = os.environ.get('CLEARS_PAGE_NAME')

source_sh = gc.open_by_key(real_clears_sheet_id)
dest_sh = gc.open_by_key(fake_clears_sheet_id)
worksheet = source_sh.worksheet(clears_page_name)
worksheet.copy_to(dest_sh.id)
```
